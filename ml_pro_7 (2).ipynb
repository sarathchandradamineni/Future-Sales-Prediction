```python
#DATA PRE-PROCESSING
```

```python
#dealing with the null values
data.isnull().sum()/len(data) * 100
```

```python
#replacing the null values of the attribute Item_Weight with median of Item_Weight
data['Item_Weight'] = data['Item_Weight'].fillna(data['Item_Weight'].median())
```

```python
#replacing the null values of Outlet_Size with mode of Outlet_Size attribute
data['Outlet_Size'] = data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0])
```

```python
#calculating varience of each attribute
data.var()
```

```python
#to find the contents of the attribute Outlet_Size
data['Outlet_Size'].value_counts()
```

```python
#to find the contents of the attribute Item_Fat_Content
data['Item_Fat_Content'].value_counts()
```

```python
#replacing 'LF','low fat' with 'Low Fat'
data['Item_Fat_Content'] = data['Item_Fat_Content'].replace(['LF','low fat'],'Low Fat')
```

```python
#replacing 'reg' with 'Regular'
data['Item_Fat_Content'] = data['Item_Fat_Content'].replace('reg','Regular')
```

```python
#to find the contents of the attribute Item_Fat_Content
data['Item_Fat_Content'].unique()
```

```python
data
```

```python
#imputing the visibility of items having Item_Visibility = 0, As item visibility = 0 makes no sense
#for this imputing these values with the mean of visibility group by item identifier
visibility_item_avg = data[['Item_Identifier', 'Item_Visibility']].groupby( 'Item_Identifier' ).mean()
visibility_item_avg.head()

def visibility_imputing(x):
    visibility = x[0]
    i = x[1]
    
    if visibility == 0:
        return visibility_item_avg['Item_Visibility'][visibility_item_avg.index == i]
    else:
        return visibility
    
#sum(data['Item_Visibility'] == 0)
    
data['Item_Visibility'] =  data[['Item_Visibility','Item_Identifier']].apply(visibility_imputing,axis=1).astype(float)

#sum(data['Item_Visibility'] == 0)
```

```python
#to find the contents of the attribute Item_Type
data['Item_Type'].value_counts()
```

```python
#creating new attribute 'Product_Type' from Item_Identifier attribute
data['Product_Type'] = data['Item_Identifier'].apply( lambda x: x[0:2] )
data['Product_Type'] = data['Product_Type'].map( {'FD': 'Food', 'NC':'Non-Consumable', 'DR':'Drinks'} )   
data['Product_Type'].value_counts()
```

```python
a = data[['Product_Type', 'Item_Outlet_Sales']].groupby( 'Product_Type' ).median()
a.plot( kind='bar', color='#B3D8E6', figsize=( 12, 7 ) );

plt.xlabel( 'Product_Type' )
plt.ylabel( 'Median Item Outlet Sales')
plt.title( 'Impact of Product_Type on Item Outlet Sales' )
plt.xticks( rotation = 0 )
```

```python
new_data_list = new_data.values.tolist()
new_data_output = new_data['Item_Outlet_Sales']
new_data_input = new_data.drop('Item_Outlet_Sales',axis = 1)
new_data_input_ndarray = new_data_input.iloc[:].values
new_data_output_ndarray = new_data_output.iloc[:].values
new_data_input_list = new_data_input_ndarray.tolist()
new_data_output_list = new_data_output_ndarray.tolist()
```

```python
from sklearn.model_selection import train_test_split
train_data_input, test_data_input, train_data_output, test_data_output = train_test_split(new_data_input_list,new_data_output_list,test_size = 0.2,random_state = 1000)
```

```python
from sklearn.linear_model import LinearRegression
lnr_regressor = LinearRegression() 

from sklearn.ensemble import RandomForestRegressor
rnd_forest_regressor = RandomForestRegressor(n_estimators=100,max_depth=6, min_samples_leaf=50,n_jobs=4)

from sklearn.naive_bayes import GaussianNB
naivebayes = GaussianNB()
```

```python
from sklearn.model_selection import StratifiedKFold, KFold
import time
skf = StratifiedKFold(n_splits=3)

kf = KFold(n_splits=10)
kf.get_n_splits(new_data_input)

def ten_fold(model):
    accuracy = []
    f1 = []
    training_time_taken=[]
    for train_index, test_index in kf.split(new_data_list):
        
        X_train, X_test = new_data_input_ndarray[train_index], new_data_input_ndarray[test_index]
        y_train, y_test = new_data_output_ndarray[train_index], new_data_output_ndarray[test_index]
        
        before = time.time()
        model.fit(X_train, y_train)
        after = time.time()

        training_time = round(after-before, 4)

        predict = model.predict(X_test)
        acc = round(model.score(X_test, y_test), 4)
        #f11 = round(f1_score(predict, y_test), 4)
        accuracy.append(acc * 100)
        #f1.append(f11)
        training_time_taken.append(training_time)
    return accuracy, training_time_taken
```

```python
lnr_reg_acc,lnr_reg_time = ten_fold(lnr_regressor)
rnd_frst_acc,rnd_frst_time = ten_fold(rnd_forest_regressor)
```
